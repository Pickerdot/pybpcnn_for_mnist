```python
t = np.array([[0.6,0.2,0.4]])
input_data = np.array([[0.4,0.1,0.2,-0.9]])
W = np.random.randn(4, 3)
b = np.random.randn(1, 3)
lr = 0.005
h_layer = ly.hidden_layer(W, b, lr)
Loss_layer = ly.Loss()
```

```python
dz_memo = []
loss_memo = []
b_memo = []
W_memo = []
for i in range (20000):
    z = h_layer.forward(input_data)
    loss = Loss_layer.forward(z,t)
    dz = Loss_layer.backward()
    dx, dy, W, b = h_layer.backward(dz)
    loss_memo.append(loss)
    b_memo.append(b[0][0])
    W_memo.append(W[0][0])
    if i % 100 == 0:
        print("W:",W)
        print("Z:", z)
        print("dz:", dz)
        print("dy:", dy)
        print("b:", b)
        
```
